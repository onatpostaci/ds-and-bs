## Loading the CSV Data

This notebook demonstrates how to load a CSV file using pandas and visualize the data.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = './XRPUSDT-1d-2023-01.csv'  # Update the path if necessary
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python

# Assuming the first column is a timestamp and the fifth column is the closing price
data['Timestamp'] = pd.to_datetime(data.iloc[:,0], unit='ms')
data['Close'] = data.iloc[:,4]

# Plotting the closing price over time
plt.figure(figsize=(12,6))
plt.plot(data['Timestamp'], data['Close'])
plt.title('Closing Price over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USDT)')
plt.grid(True)
plt.show()
```
