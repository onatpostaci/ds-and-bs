# Data Analysis and Visualization Notebook

## Loading the CSV Data

This notebook demonstrates how to load a CSV file using pandas and visualize the data.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = './XRPUSDT-1d-2023-01.csv'  # Update the path if necessary
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python

# Assuming the first column is a timestamp and the fifth column is the closing price
data['Timestamp'] = pd.to_datetime(data.iloc[:,0], unit='ms')
data['Close'] = data.iloc[:,4]

# Plotting the closing price over time
plt.figure(figsize=(12,6))
plt.plot(data['Timestamp'], data['Close'])
plt.title('Closing Price over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USDT)')
plt.grid(True)
plt.show()
```

## Binance Public Data Analysis

This notebook includes an analysis of the Binance public data, specifically the Kline data. The data represents various trading information such as open time, open price, high price, low price, close price, volume, and more.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Column names based on Binance Kline data format
column_names = ['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 
                'Close time', 'Quote asset volume', 'Number of trades', 
                'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore']

# Load the CSV file
file_path = './XRPUSDT-1d-2023-01.csv'  # Update the path if necessary
data = pd.read_csv(file_path, names=column_names)

# Convert timestamps from milliseconds to datetime
data['Open time'] = pd.to_datetime(data['Open time'], unit='ms')
data['Close time'] = pd.to_datetime(data['Close time'], unit='ms')

# Display the first few rows of the dataframe
data.head()
```

```python

# Plotting Open, High, Low, Close prices
plt.figure(figsize=(15, 7))
plt.plot(data['Open time'], data['Open'], label='Open Price')
plt.plot(data['Open time'], data['High'], label='High Price')
plt.plot(data['Open time'], data['Low'], label='Low Price')
plt.plot(data['Open time'], data['Close'], label='Close Price')
plt.title('Price Trends over Time')
plt.xlabel('Date')
plt.ylabel('Price (USDT)')
plt.legend()
plt.grid(True)
plt.show()
```
